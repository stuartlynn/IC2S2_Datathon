```python
with open('msa_dict.pkl', 'rb') as f:
    dic = pickle.load(f)
```

```python
print(len(dic), len(set(dic.values())))
```

```python
postfix = " Metro Area"
mca_to_geoid = {}
for geoid, mca_raw in dic.items():
    mca = mca_raw.split(postfix)[0]
    mca_to_geoid[mca] = geoid
```
